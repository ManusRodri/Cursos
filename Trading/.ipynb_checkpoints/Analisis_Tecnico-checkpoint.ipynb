# Que es un mercado financiero?
+ Un mercado financiero es un espacio ya sea fisico o virtual, en este espacio compradores y vendedores intercambian activos, con agentes economicos de por medio.

+ Este sistema facilita la interaccion entre aquellas personas o entidades que cuentan con el capital para invertir y las que nesecitan capital.

+ Caracteristicas de un mercado financiero

  1. Amplitud: cuenta con un volumen significativo de activos y liquidez
 
  2. Transparencia: la informacion monetaria de los activos debe ser publico para todas las personas. Puede verse desde graficos y estadisticas
 
  3. Libertad: puedes comprar y vender en cualquier momento
 
  4. Profundidad: cuantas mayores operaciones de compra / venta que hayan en el mercado, mas profundo es
 
  5. Flexibilidad: liquidez y facilidad de compra / venta
 
+ Tipos de mercados financieros:

  1. Mercados monetarios: Donde se ejecutan transacciones interbancarias, deuda publica y se maneja nuestra moneda.
 
  2. Mercados de capitales: Se refiere a los mercados de renta fija o renta variable o los derivados financieros
 
  3. Mercado de acciones: En este mercado se realiza la compra/venta de titulos de propiedad de compa√±ias publicas, por ejemplo: Apple, Tesla, Google, Meta, etc.
 
  4. Mercado de futuros: Es un acuerdo entre dos inversionistas en el que se comprometen a comprar y vender un activo, acordando en el momento en que se realice el intercambio las condiciones de dicha compraventa en el futuro.
 
  5. Mercados de materias primas: oro, plata, petroleo, etc
 
  6. Mercado de divisas (Forex):  Mercados de intercambio de divisas, por ejemplo: dolar-euro, euro-libra esterlina, etc.
 
  7. Mercado de criptodivisas: En este mercado se realiza la compra y venta de criptoactivos, el mas conocido es Bitcoin; sin embargo, existen miles de proyectos con Ether, BNB, Solana, USD Coin, etc.
 
# Que es un activo financiero?

+ es un instrumento que tiene valor o puede generar ingresos. Es un documento o titulo emitido por una empresa, banco, institucion privada o publica, cuyo comprador o propietario espera que le produzca rendimientos.

# Que es un CDF?
+ De sus siglas en ingles, un Contract For Difference es un contrato que se basa en el cambio del valor de los activos financieros mediante el cual el comprador paga al vendedor para adquirir un activo adyacente que cambiara de valor en el tiempo.
  
  * Si la diferencia es positiva, el vendedor paga la diferencia al comprador.
  
  * Si la diferencia es negativa, el comprador es quien debe pagar la diferencia al vendedor.

# Soportes y Resistencias

## Que son

- Un **soporte** es un nivel del grafico donde el precio **deja de caer** y **rebota al alza**. Representa una zona donde los compradores entran y evitan que el precio siga bajando.

- Una **resistencia** es un nivel donde el precio **deja de subir** y **gira a la baja**. Representa una zona donde los vendedores intervienen y frenan el avance del precio.

**En resumen:**  
- El **limite inferior** es el soporte.  
- El **limite superior** es la resistencia.  
- Son puntos donde el precio **se detiene temporalmente** y puede **cambiar de direccion**.

---

## Por que son importantes

- Ayudan a **anticipar zonas de giro** del mercado.  
- Sirven para **colocar ordenes de entrada, salida y stop loss** de forma mas precisa.  
- Actuan como **niveles psicologicos** que influyen en el comportamiento de los traders.  
- Un soporte roto puede convertirse en **resistencia** y una resistencia rota puede volverse **soporte**.  
- Son una base clave para estrategias de **rebotes**, **rupturas (breakouts)** y **pullbacks**.

---

## Tipos de soportes y resistencias

### 1. Dinamicos

- Se determinan con **indicadores que se mueven junto al precio**, como:
  - Medias moviles (EMA, SMA)  
  - Bandas de Bollinger  
  - Nubes de Ichimoku

- No marcan un nivel fijo: **se ajustan constantemente** al movimiento del mercado.  
- Evolucionan junto con **tiempo y precio**.  
- Son muy utiles para operar **en tendencias**.

*Ejemplo:* si el precio rebota repetidamente en la media movil de 50 periodos, esa media funciona como **soporte dinamico**.

  
+ 

---

### 2. Semi-dinamicos

- Son niveles **intermedios entre horizontales y dinamicos**.  
- Se calculan con **formulas o proporciones** que se **actualizan de forma periodica**, no en tiempo real.  
- Ejemplos:
  - Retrocesos de Fibonacci  
  - Pivot points (diarios, semanales o mensuales)  
  - Rangos calculados con promedios periodicos

- Cambian de forma **escalonada**, por ejemplo al cierre diario o semanal.  
- Se desarrollan en **tiempo y precio**, pero con menos frecuencia que los dinamicos.

*Ejemplo:* un nivel de retroceso Fibonacci 0.618 que se recalcula al final del dia actua como **zona de soporte o resistencia semi-dinamica**.

+ 

---

### 3. Horizontales

- Son **niveles fijos** donde el precio **rebota repetidamente** antes de romper.  
- No evolucionan con el tiempo: se mantienen **estables** hasta que el precio rompe el nivel.  
- Se identifican observando zonas con varios rebotes previos.  
- Se desarrolla principalmente **tiempo**, el **precio cambia** solo al romperse.  
- Suelen coincidir con **niveles psicologicos** (ej. 1000, 20000, 50000).

*Ejemplo:* si el precio toca 20000 varias veces y rebota, esa zona es una **resistencia horizontal**.



+ 

---

### 4. Diagonales

- Se trazan uniendo **maximos o minimos crecientes o decrecientes**.  
- Forman lineas **con pendiente alcista o bajista**.  
- Se dibujan **manualmente** segun la tendencia.  
- Evolucionan en el tiempo de forma **diagonal**.  
- Sirven para **identificar tendencias** y **zonas de rebote** dentro de un canal.

*Ejemplo:* en una tendencia alcista, una linea uniendo minimos ascendentes actua como **soporte diagonal**

- 

---

## Cambio de rol: Soporte ‚Üî Resistencia

- Cuando el **precio rompe un soporte**, ese nivel roto puede **convertirse en resistencia futura**.  
- Cuando el **precio rompe una resistencia**, esta puede **convertirse en soporte**.  
- Este fenomeno se llama **cambio de polaridad**.  
- Es clave para estrategias de **breakout** y **pullback**.

*Ejemplo:* si el precio rompe 1800 hacia abajo y luego vuelve a subir, 1800 puede actuar como resistencia.

---

## Consejos practicos

- Cuantos **mas toques** tenga un nivel, **mas fuerte** es.  
- Evita llenar el grafico de niveles: enfocate en los **mas relevantes**.  
- Combina soportes y resistencias con:
  - Volumen  
  - Patrones de velas japonesas  
  - Lineas de tendencia o canales
- No son puntos exactos, sino **zonas**.  
- Cuidado con los **falsos rompimientos (fakeouts)**: confirma antes de operar.

---

## Estrategias basicas

1. **Rebote**  
   - Comprar en soporte o vender en resistencia.  
   - Colocar stop loss debajo (o encima) del nivel.

2. **Ruptura (Breakout)**  
   - Operar cuando el precio rompe un nivel con fuerza.  
   - Confirmar con volumen alto o cierre fuera del rango.

3. **Pullback**  
   - Esperar que el precio rompa un nivel y vuelva a testearlo.  
   - Entrar en la direccion del movimiento principal.

---

# Terminos

## Breakout y Pullback

### Breakout (Ruptura)

Un **breakout** ocurre cuando el **precio atraviesa con fuerza** un nivel importante de soporte o resistencia.  
Este movimiento suele ir acompa√±ado de **mayor volumen**, lo que indica **fuerza en la nueva direccion**.

- Si el precio **rompe una resistencia**, puede indicar el **inicio de una tendencia alcista**.  
- Si el precio **rompe un soporte**, puede indicar el **inicio de una tendencia bajista**.

*Ejemplo:*  
Si el precio ha rebotado varias veces en 20000 y finalmente **rompe ese nivel hacia arriba** con fuerza, se considera un **breakout alcista**.

Caracteristicas comunes:
- Velas grandes y solidas.  
- Aumento de volumen.  
- Alta probabilidad de continuacion del movimiento.  
- Muchos traders aprovechan estos movimientos para **entrar en la nueva tendencia**.

---

### Pullback (Retroceso controlado)

Un **pullback** ocurre cuando, **despues de un breakout**, el precio **retrocede temporalmente** para **probar nuevamente** el nivel que rompio.  
Si ese nivel ahora **actua como soporte o resistencia inversa**, confirma la **validez de la ruptura**.

- Si rompe una resistencia y luego baja a probarla como soporte ‚Üí pullback alcista.  
- Si rompe un soporte y luego sube a probarlo como resistencia ‚Üí pullback bajista.

*Ejemplo:*  
El precio rompe 20000 hacia arriba, sube hasta 20500 y **retrocede nuevamente a 20000**.  
Si rebota en 20000 sin volver a caer, ese pullback confirma la **nueva tendencia alcista**.

Caracteristicas comunes:
- Retroceso moderado y controlado.  
- Confirma niveles importantes.  
- Ofrece una **segunda oportunidad de entrada** con menor riesgo.  
- Muy utilizado por traders para **entrar despues de la ruptura inicial**.

---

## Breakout vs Pullback

| Caracteristica         | Breakout                          | Pullback                                 |
|-------------------------|-------------------------------------|-------------------------------------------|
| Momento                | Justo cuando se rompe un nivel     | Despues de la ruptura                    |
| Direccion              | Movimiento fuerte inicial          | Retroceso controlado                     |
| Objetivo               | Aprovechar la fuerza del impulso  | Confirmar la validez del nivel roto      |
| Riesgo                 | Mayor (entrada en movimiento rapido) | Menor (entrada tras confirmacion)      |
| Estrategia             | Entrar en la ruptura               | Entrar en el rebote tras la ruptura     |

# Como dibujar soportes y resistencias

## 1. Soportes y resistencias dinamicos
- No se dibujan con lineas; el grafico los muestra solo usando medias moviles como la EMA20, EMA50 o EMA200.  
- Estas lineas cambian junto con el precio y ayudan a ver si el precio rebota (soporte) o se frena (resistencia).  
- Cuanto mas veces el precio toque una media y rebote, mas fuerte es ese nivel.

---

## 2. Soportes y resistencias horizontales
- Son los niveles mas usados. Se dibujan con lineas o zonas planas donde el precio se ha detenido o rebotado varias veces.  
- Como hacerlo:
  1. Mira el grafico y busca los puntos donde el precio se detuvo o cambio de direccion.  
  2. La mejor forma de ubicar la linea es empezar donde el precio esta cotizando actualmente, y luego ajustarla para que coincida con los ultimos rebotes o frenos (OJO es peligroso)
  3. Otra forma es buscar las barras poderosas, para las alcistas se pone soportes (se pone abajo de la barra) y las bajista como resistencia (se pone arriba de la barra)
  4. Dibuja una linea horizontal que pase por esos puntos (mechas o cierres de velas).  
  5. Cuantos mas toques tenga, mas importante sera.  
- Ejemplo: si el precio sube, toca varias veces un mismo nivel y baja, ahi tienes una resistencia. Si baja, toca varias veces y sube, ahi tienes un soporte.

---

## 3. Soportes y resistencias diagonales
- Se trazan en forma de linea inclinada.
- Como hacerlo:
  1. Si el precio va subiendo, une los minimos mas bajos (se buscan los ultimos 2 o 3 minimos) ‚Üí soporte diagonal.  
  2. Si el precio va bajando, une los maximos mas altos (se buscan los ultimos 2 o 3 maximos) ‚Üí resistencia diagonal.  
  3. Cuantos mas puntos toque esa linea sin romperla, mejor funciona.
- Sirven para ver la direccion de la tendencia y posibles rompimientos.

---

## 4. Soportes y resistencias semi-dinamicos
- No son fijos ni cambian todo el tiempo, sino que se ajustan poco a poco con el movimiento del precio.  
- Ejemplos:
  - Los ultimos picos o valles (puntos recientes donde el precio reboto fuerte).  
  - Los niveles donde hubo mucho movimiento o pausa antes de un cambio.  
- Como dibujarlos:
  1. Marca los ultimos rebotes fuertes o zonas de pausa.  
  2. Actualiza esos niveles cuando el grafico cree nuevos puntos importantes.

---

## 5. Consejos rapidos
- Empieza con los niveles horizontales (los mas faciles de ver).  
- Luego agrega lineas diagonales para marcar la tendencia.  
- Usa las medias moviles para confirmar zonas dinamicas.  
- No llenes el grafico de lineas: pocos niveles, pero claros.  
- Cuantas mas veces el precio respete un nivel, mas fuerte sera.
