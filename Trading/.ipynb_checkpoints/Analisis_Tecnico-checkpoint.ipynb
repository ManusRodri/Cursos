# Que es un mercado financiero?
+ Un mercado financiero es un espacio ya sea fisico o virtual, en este espacio compradores y vendedores intercambian activos, con agentes economicos de por medio.

+ Este sistema facilita la interaccion entre aquellas personas o entidades que cuentan con el capital para invertir y las que nesecitan capital.

+ Caracteristicas de un mercado financiero

  1. Amplitud: cuenta con un volumen significativo de activos y liquidez
 
  2. Transparencia: la informacion monetaria de los activos debe ser publico para todas las personas. Puede verse desde graficos y estadisticas
 
  3. Libertad: puedes comprar y vender en cualquier momento
 
  4. Profundidad: cuantas mayores operaciones de compra / venta que hayan en el mercado, mas profundo es
 
  5. Flexibilidad: liquidez y facilidad de compra / venta
 
+ Tipos de mercados financieros:

  1. Mercados monetarios: Donde se ejecutan transacciones interbancarias, deuda publica y se maneja nuestra moneda.
 
  2. Mercados de capitales: Se refiere a los mercados de renta fija o renta variable o los derivados financieros
 
  3. Mercado de acciones: En este mercado se realiza la compra/venta de titulos de propiedad de compa√±ias publicas, por ejemplo: Apple, Tesla, Google, Meta, etc.
 
  4. Mercado de futuros: Es un acuerdo entre dos inversionistas en el que se comprometen a comprar y vender un activo, acordando en el momento en que se realice el intercambio las condiciones de dicha compraventa en el futuro.
 
  5. Mercados de materias primas: oro, plata, petroleo, etc
 
  6. Mercado de divisas (Forex):  Mercados de intercambio de divisas, por ejemplo: dolar-euro, euro-libra esterlina, etc.
 
  7. Mercado de criptodivisas: En este mercado se realiza la compra y venta de criptoactivos, el mas conocido es Bitcoin; sin embargo, existen miles de proyectos con Ether, BNB, Solana, USD Coin, etc.
 
# Que es un activo financiero?

+ es un instrumento que tiene valor o puede generar ingresos. Es un documento o titulo emitido por una empresa, banco, institucion privada o publica, cuyo comprador o propietario espera que le produzca rendimientos.

# Que es un CDF?
+ De sus siglas en ingles, un Contract For Difference es un contrato que se basa en el cambio del valor de los activos financieros mediante el cual el comprador paga al vendedor para adquirir un activo adyacente que cambiara de valor en el tiempo.
  
  * Si la diferencia es positiva, el vendedor paga la diferencia al comprador.
  
  * Si la diferencia es negativa, el comprador es quien debe pagar la diferencia al vendedor.

# Analisis tecnico

+ El analisis tecnico es el estudio de los movimientos de precio de un activo financiero, con el objetivo de hacer proyecciones a partir de graficos y de esa forma poder tomar mejores decisiones de inversion.

+ El analisis tecnico tiene un cierto margen de error, pero el conjunto de herramientas y estrategias nos pueden ayudar a tomar mejores decisiones y disminuir los riesgos. Tambien es importante tener en cuenta el gestionar y cuidar nuestro capital, manejar y gestionar de manera correcta nuestras emociones y estudiar y analizar los diferentes indicadores.

+ Filosofia del analisis tecnico:

  * Los movimientos del precio lo cuentan todo.
 
  * Los precios se mueven por tendencias.
 
  * La historia se repite.

# Analista tecnico

+ Especula a partir de la fluctuacion de los precios de un activo financiero de forma profesional.

## Tipos (Estos son los mas principales):

- **Scalper**  
  - **Temporalidad:** Segundos a minutos  
  - **Tiempo de operaciones:** Muy corto plazo  
  - **Explicacion:** Realiza muchas operaciones en el mismo dia buscando pequenas ganancias acumuladas  

- **Day trader (Intradia)**  
  - **Temporalidad:** Minutos a horas  
  - **Tiempo de operaciones:** Dentro del mismo dia  
  - **Explicacion:** Aprovecha la volatilidad diaria y nunca deja posiciones abiertas de un dia a otro  

- **Swing trader**  
  - **Temporalidad:** Dias a semanas  
  - **Tiempo de operaciones:** Mediano plazo  
  - **Explicacion:** Mantiene posiciones varios dias para aprovechar oscilaciones del mercado  

- **Position trader**  
  - **Temporalidad:** Semanas a meses o a√±os  
  - **Tiempo de operaciones:** Largo plazo  
  - **Explicacion:** Se basa en analisis fundamental y macroeconomico para seguir grandes tendencias  

### Ventajas y desventajas:

#### **Scalper**

* **Ventajas**

  * Muchas oportunidades de entrada y salida
  * Acumula pequenas ganancias que se vuelven significativas
  * Baja exposicion a grandes movimientos en contra

* **Desventajas**

  * Alta presion y estres por la velocidad
  * Requiere atencion constante
  * Comisiones y spreads reducen rentabilidad

---

#### **Day trader (Intradia)**

* **Ventajas**

  * Se aprovecha de la volatilidad diaria
  * Resultados rapidos al final del dia
  * Mayor flexibilidad en gestion de capital

* **Desventajas**

  * Requiere muchas horas frente a la pantalla
  * Necesita reaccion rapida y control emocional
  * Alta exposicion a noticias intradia

---

#### **Swing trader**

* **Ventajas**

  * Menor demanda de tiempo
  * Aprovecha movimientos amplios
  * Menores comisiones que scalping

* **Desventajas**

  * Riesgo de mantener posiciones abiertas de noche o fin de semana
  * Requiere paciencia y disciplina
  * Noticias inesperadas pueden cambiar la tendencia

---

#### **Position trader**

* **Ventajas**

  * Menos operaciones, menos estres
  * Puede obtener grandes ganancias en tendencias
  * Requiere poco monitoreo constante

* **Desventajas**

  * Alta exposicion a riesgos macro o politicos
  * Capital inmovilizado por largo tiempo
  * Dificil mantener confianza en retrocesos prolongados
 
---

# Plataformas de inversion

+ Un Broker, plataforma de intercambio o inversion, es un programa informatico que te permite realizar operaciones con instrumentos financieros en tiempo real. Estas entidades funcionan como intermediarios, ejecutando las ordenes del mercado bursatil, y el trader es quien opera la plataforma.

## Que es un broker financiero

- **Broker**
  - Es una entidad o empresa financiera que:
    - Ejecuta ordenes de compra y venta de activos en los mercados bursatiles  
    - Cobra comisiones por este servicio  
    - Cuenta con licencia para la compra y venta de acciones  
    - Es necesario para que los traders puedan operar en dichos mercados  

- **Diferencia entre broker y trader**
  - Los brokers proveen la plataforma y los medios para realizar operaciones en el mercado  
  - Los traders utilizan esa plataforma para efectuar compras y ventas  
  - Los traders no pueden operar directamente en el mercado sin un broker
