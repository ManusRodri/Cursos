# Historia

1. Historia Lineal

   + Mesopotamia:
       + Quedaba en lo que hoy conocemos como: Irak, Turquia y Siria. Es el lugar donde todo empez.
    
       + 2250-2000 BC: inicia la renta de las tierras.
         
       + Los sumerios manejan las tierras, hay superavit de tierras con acceso a aguas.
         
       + 2000 BC: la tercera dinastia Ur recibe impuestos (gun-mada) en forma de renta, remanentes de la produccion obtenida.
         
       + De no haber remanentes se iba a la guerra entre vecinos o se optaba por adquirir prestamos (usura), dando origen a los intereses.
         
       + 1754 BC: surge el Codigo de Hammurabi con las primeras leyes contractuales y de comercio, se habla ya de deudas y se amplia la practica de la usura.
    
   
   + Grecia
       + En el siglo IV AC, en la antigua Grecia surge el libro "Oeconomicus" donde se hacen preguntas para llegar a conclusiones propias con relacion a la contratacion de granjeros (no esclavitud) y surgen los primeros 'manager' o administradores de tierras.
    

       + Surge otro libro: "Alegorias de la granja, de Grecia a Roma" donde se inicia a comprar tierras con problemas, para despues trabajarlas y las vendian con alguien interesado (asi surgen los margenes de ganancia). Nota: House flipping: Es la actividad de comprar propiedades arruinadas, se trabajan y se mejoran, para luego venderlas con un margen atractivo (lo que hoy conocemos como inversionistas de bienes raíces).
    

    + Roma (ya una república bien establecida)
        + En el año 200 AC, se consolidad la inversion en bienes raices, haciendo ricas a muchas familias de la elite romana.

2. Historia del `interes`
   + El interes es el motivo por el que una persona o una institucion esta interesada en prestarte dinero
      
   + En egipto 300 BC, las tasas de interes eran del 100%, no fue hasta Ptolomeo (post 30BC) que las tasas de interes se igualaron con las del naciente imperio Romano en un 25%

   +  Roma fue financiando mucho de su expansion, gracias al movimi9ento prestamo-interes de la clase noble para financiar campañas de conquista
  
   +  Durante la edad Media y el Renacimiento, el fenomeno usurero (prestamos con intereses muy altos) gano popularidad clandestina, pero tambien perdio buena reputacion
  
   +  De todas las categorias o tipos de inversion el comercio (trade) era el mas riesgoso y a la vez el que ams retorno podia ofrecer
  
   +  La figura de la corporacion surgio como una entidad que protegia a los inversionistas que pagaban la expedicion, esta acuataba como escudo.
  
   +  La primera corporacion que existio era The British East India Company que fue fundada en 1600 y cerro sus puertas despues de 1800, fue una de las pioneras en la entidad corporativa
  
   +  Esta empresa fue la primera que constituyo el modelo de ownership (propiedad privada) a traves de acciones (shares of stock) y la primera que produjo estructuras administrativas y legales como las de hoy en dia
  
   +  The British East India Company fue tambien la primera en usar un sistema de proteccion contra responsabilidad legal, protegiendo a sus inversionistas. Asi nacio la responsabilidad limitada (Limited Liability)
  
   +  Las corporaciones se convirtieron en el vehiculo ideal para invertir y financiar en las riesgosas expediciones
  
   +  En 1602, nacio The Dutch East India Company, esta se convirtio en la primera empresa "publica", poniendo a disposicion del publico inversionista sus acciones en el Amsterdam Stock Exchange
  
3. Breve historia de la inversion
   
   + En un principio la compra-venta de acciones publicas en los primeros Stock Exchange de Europa (Amsterdam, Belgica) sucedia en los cafes y en algunas plazas publicas
  
   + La primera bolsa de valores que emitio de manera formal acciones al publico similar a como se hacia con emisiones de deuda de gobierno fue la bolsa de valores de Londres
  
   + El London Stock Exchange fue creado en 1801 y rapidamente el New York Stock Exchange fue inaugurado en 1817 y se convirtio en la Bolsa mas importante y poderosa del Mundo
  
   + Las Bolsas de Valores del mundo encuentran su razon de existir en el concepto de 2 temas: el manejo de riesgos y como la corporacion es una entidad que nacio para protegerse del riesgo, y dos la busqueda de desarrollo economico y de negocios.
   

# Categorias y Tipos de inversion

+ Bienes inmuebles (Real State)
+ Productos Bancarios (Bank Products)
+ Bonos (Bonds)
+ Acciones (Stocks, ETFs)
+ Fondos de inversion y de Retiro / Pensiones (Mutual Funds, Pension / Retirement Funds)
+ Seguros (Insurance)
+ Opciones y Futuros (Options & Futures)
+ Otros productos (Alternative Investments)



1. Bienes Inmuebles(Real state): Inversiones en tierra o en los edificion construidos sobre ella y/o el ecosistema natural habitado dentro de sus limites

   
   + Puede ser comercial o Residencial (Comprando propiedades o invirtiendo en instrumentos relacionados)
  
   + Se genera valor por medio de:
     
     *  Cash-flow / renueve con rentas (arriendo)
       
     *  Apreciacion del inmueble o accion sobre el paso del tiempo
       
   + Los factores locales/geograficos pueden afectar de manera positiva o negativa y tambien afecta el mercado del sitio

2. Productos bancarios(Bank Products): Productos adquiridos de la banca. Son generalmente liquidos (facilidad con la que un activo puede ser convertido en dinero efectivo sin afectar significativamente su valor), pero ofrecen un interes muy bajo
   
   + Algunos son:
  
     * Cuentas de ahorro: Alta liquidez, bajo costo pero bajo interes

     * Certificados de deposito: Mayor interes que las cuentas de ahorro pero menos liquidos y tienen un plazo de tiempo determinado
    
     * Cuentas de Mercado de Dinero: Cuentas de deposito similar a la cuenta de ahorro pero con mayor interes y generalmente con mayores restricciones administrativas

3. Bonos(Bonds): Instrumento de deuda donde un inversionista le hace un prestamo a una institución publica/agencia federal/institucion privada, a cambio de un pago de interes en un cierto plazo mas el repago del capital a la fecha de vencimiento del banco


+ 

+ Acciones (Stocks, ETFs): Son las inversiones que representan la propiedad proporcional de una entidad corporativa o empresa publica. Las acciones son la division de titulos de propiedad de una entidad corporativa o empresa publica

+ La manera en la que los inversionistas obtienen su retorno de inversion (o una perdida) es a traves de dos medios:

  * Dividendos de la empresa en caso de que los distribuya (los dividendos son una parte de las ganancias totales,a cada inversionista o socio se le da porcion de un porcentaje de una parte de las ganancias de la empresa)
 
  * Apreciacion o Despreciacion de el valor de cada accion en el mercado tras una transaccion de venta o compra (Capital Gains)

+ El valor de la accion de una empresa depende de varios factores que hacen que la inversion en el mercado bursatil de acciones (en cualquier parte del mundo) sea uno de los instrumentos mas volatiles y riesgosos:

  * Ambientes de mercado en general (politica, economia, ambiente social)
    
  * Ganancias de la empresa o del sector (ETFs) al que pertenece la empresa
    
  * Demanda o "Hype" del sector o de la empresa (IPO, FANG)
    
  * Factores "irracionales" o completamente random (un tweet de su CEO, un discurso de un funcionario publico de gran importancia internacional)
 
+ Fondos de inversion y Retiro: Vehiculo de inversion en el cual se agrupa a varios inversionistas con un fin o estrategia de inversion especifica y generalmente dejando su administracion y diversificacion en manos de un "Fund Manager"

+ Existen gran variedad de tipos de fondo de inversion:

  * Mutual Funds (Stocks, Bonds, Money Market, Open/Close, etc)
    
  * Exchange Traded Funds (ETFs)
    
  * Venture Capital Funds
    
  * Fondos de Fondos (Fondos Soberanos)
    
  * Fondos de Pension (Fondos de Retiro)
    
  * Hedge Funds (no registrados ante la SEC en Estados Unidos)

+ Insurance(Seguros de Vida u otros tipos de seguros): aunque no sea comun considerarlos como una inversion, su popularidad esta aumentando como estrategia de inversion.

- Ventajas:
  - Construye valor sobre el Premium (interes compuesto) con ventajas fiscales
  - Es una buena estrategia de "Estate Planning" (Planificacion Patrimonial)
  - Los Premium nunca incrementan
  - Proteccion tradicional contra Muerte inesperada
  - Sus dividendos pueden cubrir el Premium en cierto plazo
  - Obliga a ahorrar

- Desventajas:
  - Premium considerablemente mas alto que el Seguro de Vida de Termino (Term)
  - Poco flexible (La estructura de pago de premiums y las condiciones no cambian)
  - La compania que lo emite (Aseguradora) tiene completo control sobre el Premium

+ Derivados: Instrumentos de inversion cuto valor esta basado en el precio o valor de otro activo subyacente

+ Algunos son:

  * SWAP (Intercambios de cahs-flow, de intereses / deuda, moneda y precios flotantes)
  * Opciones (Contratos de derecho de compra-venta sobre stocks, basicos, etc)
  * Futuros (Contratos de compra o venta futura de un activo subyacente)
  * CFDs (Contrato por Diferencia de compra o venta donde el que compra/vende -segun si el valor se aprecia o se desprecia- paga la diferencia de precio del activo subyacente al vencimiento de este)
 
- Vehiculos alternativos de inversion: son mas complejos y poco comunes. Ayudan a la diversificacion de inversionistas sofisticados

- Algunos ejemplos son:

  - Notas Convertibles, Notas Reversibles, Safe Notes (Usados por Angels/Startups)
  - Mutual Funds Alternativos (Alt-Mutual Funds, con inversiones poco tradicionales)
  - Notas de Deuda con Principal Protegido
  - Inverse Exchange Traded Funds
  - Bonos de Catastrofe y otros vehiculos relacionados a eventos azarosos
  - ICOs y Criptomonedas (Bitcoin, etc)
