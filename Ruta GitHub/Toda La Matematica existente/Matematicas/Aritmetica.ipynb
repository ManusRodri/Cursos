# Suma

$$ a + b = c$$



# Resta
+ Operacion matematica que representa la operacion de eliminacion de objetos de una coleccion
$$ a - b = c$$



# Multiplicacion
+ operacion matematica que consiste en hallar el resultado de sumar un numero tantas veces como indique otro
$$ a*b = c $$



# Division
+ operacion matematica que consiste en separar una cantidad en partes iguales
$$ a/b = c $$

# Reglas Basicas
## Suma
- Todo numero sumado por cero es igual al mismo numero

$$ a + 0 = a $$

## Resta
- Todo numero restado por cero es igual al mismo numero

$$ a - 0 = a $$

## Multiplicacion
- Todo numero multiplicado por cero es igual a cero

$$ a * 0 = 0 $$

- Todo numero multiplicado por uno es igual a uno

$$ a * 1 = a $$

## Division
- Todo numero dividido por el mismo numero es igual a uno

$$ a / a = 1 $$

- Todo numero dividido por uno es igual al mismo numero

$$ a / 1 = a $$

# Fracciones
+ Una fraccion es un numero , que se obtiene de dividir un entero en partes iguales, esto sirve para representar decimales, y representar enteros



$$ \frac{a}{b} $$

+ E.g
$$ \frac{2}{4}$$

## Fracciones meores que 1 (Propias)
+ Una fraccion menor que uno es aquella cuyo valor numerico **NO** excede 1

```python
import matplotlib.pyplot as plt

# Datos para 4 porciones iguales; resaltamos 2/4
representacion = [2/4] * 4

# Colores: 2 en azul (mitad) y 2 en gris
colors = ['tab:blue'] * 2 + ['tab:gray'] * 2
labels = [f'Sector {i+1}' for i in range(len(representacion))]

# Dibujar gráfico
plt.figure()
plt.pie(
    representacion,
    labels=labels,
    colors=colors,
    startangle=90,
    wedgeprops={'linewidth': 2, 'edgecolor': 'black'}
)
plt.axis('equal')
plt.title('Fracción 2/4')
plt.show()
```

## Fracciones mayores que 1 (impropias)
+ Una fraccion mayor que uno es aquella cuyo valor numerico excede 1
$$ \frac{a}{b} $$
$$ a > b $$

+ E.g

$$ \frac{7}{4} $$

```python
import matplotlib.pyplot as plt

# Configuración común
sizes = [1, 1, 1, 1]          # cuatro cuartos
labels = ['1/4'] * 4
wedgeprops = {'linewidth': 2, 'edgecolor': 'black'}  # bordes negros

# --- Gráfico 1: 4/4 (un entero) ---
fig1, ax1 = plt.subplots()
ax1.pie(
    sizes,
    labels=labels,
    colors=['tab:blue'] * 4,   # todo en azul
    startangle=90,
    counterclock=False,
    wedgeprops=wedgeprops
)
ax1.set_title('Fracción 4/4 (un entero)')
ax1.axis('equal')  # círculo perfecto

# --- Gráfico 2: 3/4 ---
fig2, ax2 = plt.subplots()
ax2.pie(
    sizes,
    labels=labels,
    colors=['tab:blue'] * 3 + ['tab:gray'],  # tres cuartos azules y un cuarto gris
    startangle=90,
    counterclock=False,
    wedgeprops=wedgeprops
)
ax2.set_title('Fracción 3/4')
ax2.axis('equal')

plt.show()
```

## Fracciones en la recta numerica
### Fundamento Conceptual
+ La recta numerica es la representacion geometrica del conjunto de los numeros reales \$\mathbb{R}\$, estableciendo una correspondencia entre cada punto de la recta y un unico numero real.
+ Las fracciones, como expresiones de numeros racionales \$\mathbb{Q}\$, forman un subconjunto denso de los numeros reales y, por lo tanto, cada fraccion \$\frac{a}{b}\$ (donde \$a\$, \$b\$ son enteros y \$b \neq 0\$) tiene una ubicacion unica y precisa en esta recta.
+ Esta representacion visualiza el orden inherente al conjunto de los numeros racionales.

---
### Metodologia de Representacion

Para localizar un numero racional expresado como \$\dfrac{a}{b}\$ en la recta numerica, se sigue un procedimiento basado en la interpretacion de sus componentes:

* **El Denominador (\$b\$)**: Determina la particion del intervalo unitario. Cada intervalo de longitud $1$ en la recta (por ejemplo, $[0,1]$, $[1,2]$, $[-1,0]$, etc.) se divide en $|b|$ segmentos de igual longitud. La longitud de cada uno de estos segmentos elementales es $\dfrac{1}{|b|}$.

* **El Numerador (\$a\$)**: Indica la cantidad de segmentos elementales (de longitud $\dfrac{1}{|b|}$) que deben recorrerse a partir del origen (el punto correspondiente al $0$). La direccion del recorrido depende del signo del numerador (y/o denominador, segun la convencion de signo adoptada para la fraccion): positivo hacia la derecha, negativo hacia la izquierda.

* **Localizacion del Punto**: El punto final alcanzado despues de recorrer $|a|$ segmentos elementales en la direccion apropiada corresponde exactamente al numero racional

$\dfrac{a}{b}$

---

+ E.g

$$ \frac{3}{5} $$


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Configuracion de las fracciones
denominator = 5
target_numerator = 3  # Queremos ilustrar 3/5
ticks = [n / denominator for n in range(denominator + 1)]  # 0, 1/5, ..., 1

# Crear figura
fig, ax = plt.subplots(figsize=(10, 2))
ax.hlines(0, 0, 1.05, linewidth=2)  # recta principal

# Dibujar ticks y etiquetas
for n, x in enumerate(ticks):
    # Tick vertical
    ax.vlines(x, -0.05, 0.05, linewidth=2)
    # Etiqueta de la fraccion
    if n == 0:
        label = "0"
    elif n == denominator:
        label = "1"
    else:
        label = f"{n}/{denominator}"
    ax.text(x, -0.15, label, ha="center", va="center")

# Arcos para mostrar los saltos hasta 3/5
for step in range(1, target_numerator + 1):
    # Calcular posicion del centro del arco entre (step-1)/5 y step/5
    x_start = (step - 1) / denominator
    x_end = step / denominator
    x_center = (x_start + x_end) / 2
    width = x_end - x_start
    # Dibujar arco
    arc = Arc((x_center, 0.08), width=width, height=0.12, theta1=0, theta2=180, linewidth=2)
    ax.add_patch(arc)

# Resaltar la fraccion 3/5
x_target = target_numerator / denominator
ax.plot(x_target, 0, marker="o", markersize=6)
ax.text(x_target, 0.12, f"{target_numerator}/{denominator}", ha="center")

# Ajustes finales
ax.set_xlim(-0.02, 1.05)
ax.set_ylim(-0.25, 0.25)
ax.axis("off")
plt.title("Fraccion 3/5 en la recta numerica")
plt.tight_layout()
plt.show()
```

## Tipos de fracciones
### Fracciones propias:
+ Son las fracciones que se utilizan para representar menos de una unidad, este tipo de fracciones se pueden reconecer facilmente ya que el numerador (\$a\$) es menor al denominador (\$b\$)

$$ \frac{a}{b} $$
$$ a < b $$

+ E.g

$$ \frac{3}{5} $$
$$ \frac{4}{8} $$

### Fracciones impropias
+ Son las fracciones que se utilizan para representar mas de una unidad, este tipo de fracciones se pueden reconocer facilmente ya que el numerador (\$a\$) es mayor al denominador (\$b\$)

+ E.g

$$ \frac{12}{5} $$
$$ \frac{16}{8} $$

### Fracciones unitarias
+ Son las fracciones que representan exactamente una unidad, este tipo de fracciones se puede reconocer facilmente ya que el numerador (\$a\$) es igual al denominador (\$b\$)

$$ \frac{12}{12} $$
$$ \frac{8}{8} $$

### Fracciones decimales
+ Son las fracciones cuyo denominador es una potencia de 10

$$ \frac{12}{100} $$
$$ \frac{8}{10} $$



## Tipos de fracciones al operarlas
### Fracciones homogeneas
+ Las fracciones homogeneas son aquellas fracciones que estan divididas en la misma catidad y que tienen el mismo denominador al operarlas

$$ \frac{2}{4} + \frac{7}{4} $$

### Fracciones heterogeneas
+ Las fracciones heterogeneas son aquellas fracciones que estan divididas en diferentes catidades y que diferente denominador al operarlas

$$ \frac{2}{4} + \frac{7}{14} $$
