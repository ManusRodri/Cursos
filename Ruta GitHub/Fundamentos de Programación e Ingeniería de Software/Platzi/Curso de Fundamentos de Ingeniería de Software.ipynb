# Procesos de arranque y encendido de dispositivos:

+ Al prender tu computadora o dispositivo movil, una seal electrica viaja hacia la `tarjeta madre`, esta electricidad, que es una onda, el dispositivo lo interprera en pulsos altos y pulsos bajos, un pulso alto es 1 y un pulso bajo es 0, a esto se le conoce como `los Bits` -> unidad minima de informacion.



+ Esta seal se envia hacia la tarjeta madre, la cual tiene un chip especial de arranque que recibe la seal, en este chip se encuentra un sistema fundamental llamado:

    * `BIOS` (Basic Input Output System) -> nombre antiguo.
    * `UEFI` (Unified Extensible Firmware Interface) -> nombre nuevo

+ En dispositivos moviles, como celulares, el chip encargado es el `Boot Loader`, y el proceso que ejecuta es diferente acorde al OS (Operative System)

    * Andriod -> `PBL` (Primary Boot Loader)
    * Iphone -> Secure Room, este luego carga el iBoot

+ Estos chips reciben la seal electrica y ejecutan un procedimiento interno, el cual esta grabado en el hardware de los transistores del chip. Este proceso se llama `POST` (Power On Self Test), este proceso lo que hace es que comprueba que todo en la computadora este funcionando correctamente y este conectado.

+ Cualquier sonido inusual al incio indica un fallo detectado por el POST. Si todo esta en orden, lo siguiente a proceder es buscar en la unidad de memoria permanente Las instrucciones de arranque del OS.

+ El codigo esta organizado en ceros `(0)` y unos `(1)`, con un lenguaje que lo entiende la `CPU` (Central Processing Unit), este componente es le encargado de realizar todos los calculos matematicos que se realizan para poder generar: imagenes, sonido, videos o procesos fundamentales o acciones que uno quiera realizar.

+ En computadoras o en dispositivos de gran tamao, la CPU es un chip que se puede cambiar y es de gran tamao, algunos ejemplos de chips son:

    * Intel Core Ultra 9 285K
    * AMD Ryzen 9950X3D

+ En celulares o dispositivos con un tamao reducido, los chips se encuentran soldados y no se pueden cambiar, alrededor integra otros componentes de ese chip, a esto se le conoce como: `SoC` (System on a Chip), algunos ejemplos de chips son:

    * Qualcomm Snapdragon 8 Gen 3
    * Apple A19 Pro
    * Google Tensor G5

+ El lenguaje de ceros y unos, que sale de la unidad de memoria permanente y llegas a la CPU, instrucciones las cuales inician a ejecutar diferentes procesos, esas instrucciones nativas que interpreta la CPU, es un lenguaje que se conoce como `Assembler`.

+ La CPU ejecuta el `codigo de arranque del sistema operativo`, enviando las estructuras de informacion de los componentes del dispositivo y el uso de programas o aplicaciones a una memoria temporal de muy alta velocidad, que se conoce como `Memoria RAM` (Random Access Memory), esas estructuras se van aramando, como si se estuviara armando una ciudad desde cero a partir de sus planos.

+ El corazon del OS se conoce como `Kernel`, es el codigo en Software qe interpreta la CPU y estructura la memoria RAM para funcionar como un intermediario entre el `Software` y el `Hardware`, cada vez que tu haces una accion, el Kernel traduce las ordenes dadas por esa accion, para que esas acciones pasen de la computadora a la realidad.

+ El Kernel luego arranaca los procesos de seguridad del OS, de esa forma es posible que una computadora tenga datos con multiples dueos, pero que solo puedan se accedidos por su dueo, este es el `sistema de Usuarios y Contraseas` que crea llaves de acceso que descifran los datos privados del usuario

+ El Kernel tambien busca en los archivos del OS, el proceso de arranque de las pantallas, las pantallas son matrices de puntos luminosos llamados `Pixeles`.

+ El manual de los elementos integrados, como los son: mause, teclado, pantalla, wifi, trackpad, etc. Este tipo de manual se expresa en software  y ese software se conoce como `Drivers`.

+ Al ingresar tu usuario y contrasea y le das enter, eso envia una seal electrica a la CPU, la CPU ecuentra la orden del OS de ir a buscar en la unidad de memoria permanente un `proceso criptografico matematico` donde esa contrasea descifra la `llave de acceso` al Kernel a tus datos privados.
